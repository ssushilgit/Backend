### Cosine Similarity

```python
# Manually calculate cosine similarity
def cosine_similarity_manual(pt):
   
    # Manually calculate dot product
    dot_product = np.array([[np.sum(pt[i] * pt[j]) 
            for j in range(pt.shape[0])] 
                    for i in range(pt.shape[0])])

    # Manually calculate normalization
    norm = np.sqrt(np.sum(pt**2, axis=1))
    norm_matrix = np.outer(norm, norm)
    
    # Calculate cosine similarity scores
    similarity_scores = dot_product / norm_matrix
    
    return similarity_scores
```

```python
import numpy as np

# Define your matrix here
matrix = np.array(pt)

# Call the cosine_similarity_manual function
similarity_scores = cosine_similarity_manual(matrix)

# Print the result
print(similarity_scores)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    #index_fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[0:10]
    
    data=[]
    for i in similar_items:
        item=[]
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        if not temp_df.empty:
            # Retrieve book information
            book_title = pt.index[i[0]]
            similarity_score = i[1]
            book_author = temp_df.drop_duplicates('Book-Title')['Book-Author'].values[0]
            image_url = temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values[0]

            # Append book information to data list
            item.append(book_title)
            item.append(similarity_score)
            item.append(book_author)
            item.append(image_url)

            data.append(item)

        
    return data
```

### KNN


```python
import numpy as np

# Function to compute Euclidean distance between two vectors
def euclidean_distance(vector1, vector2):
    return np.sqrt(np.sum((vector1 - vector2)**2))

# Compute pairwise distances between rows using Euclidean distance
num_rows = pt.shape[0]
distance_matrix = np.zeros((num_rows, num_rows))

for i in range(num_rows):
    for j in range(i+1, num_rows):
        distance = euclidean_distance(pt.iloc[i].values, pt.iloc[j].values)
        # Since the distance matrix is symmetric, we can assign the same value for distance_matrix[j, i]
        distance_matrix[i, j] = distance
        distance_matrix[j, i] = distance

# distance_matrix[i, j] will give the Euclidean distance between row i and row j
```

```python
def recommend_knn(book_name):
    # index_fetch
    book_id = np.where(pt.index == book_name)[0][0]
    
    # Get distances from the precomputed distance matrix for the given book_id
    distances = distance_matrix[book_id]

    # Sort distances and get indices of the k nearest neighbors
    suggestions = np.argsort(distances)[:8]

    datas = []
    for idx in suggestions:
        items = []
        temp_dff = books[books['Book-Title'] == pt.index[idx]]
        items.extend(list(temp_dff.drop_duplicates('Book-Title')['Book-Title'].values))
        items.extend(list(temp_dff.drop_duplicates('Book-Title')['Book-Author'].values))
        items.extend(list(temp_dff.drop_duplicates('Book-Title')['Image-URL-M'].values))
        items.append(distances[idx])  # Add distance to the list of items

        datas.append(items)
        
    return datas
```
